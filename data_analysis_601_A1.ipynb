```python
import matplotlib.pyplot as plt
import pandas as pd

file_path = "Student_Grades.csv"
data = pd.read_csv(file_path)

print("Dataset Preview")
print(data.head())

#1
average_score = data.iloc[:,1:].mean()
print("\n avg score by subject")
print(average_score)


average_score.plot(kind = "bar", color ="blue", title = "average score by subject")

plt.xlabel("subject's")
plt.ylabel("average score")

#2
data['Total'] = data.iloc[:,1:].sum(axis=1)
print("\n total score")
print(data[['Student Name','Total']])
data.plot(x = 'Student Name', y = 'Total', kind = 'bar', color='blue' , title = 'total performance')
plt.ylabel("total")
plt.show

#3
data.set_index('Student Name').iloc[:, :-1].T.plot(kind='line', title="subject perfromance by student")
plt.xlabel("subjects")
plt.ylabel("score")
plt.show()




```
